```python
from ddpm import data
from ddpm import config
from ddpm.data import CIFAR10_CLASSES

import matplotlib.pyplot as plt
import numpy as np
import torchvision
```

```python
config = config.DDPMConfig()
```

```python
train_loader, test_loader = data.get_cifar10_dataloaders(batch_size=config.batch_size)
```

```python
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


dataiter = iter(train_loader)
images, labels = next(dataiter)

imshow(torchvision.utils.make_grid(images[:5]))
imshow(torchvision.utils.make_grid(images[5:10]))
```
